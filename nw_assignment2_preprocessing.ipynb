Ames, Iowa Housing Dataset


```python
import pandas as pd
pd.set_option("display.max_columns", 100)
url = "https://raw.githubusercontent.com/cs6220/cs6220.spring2019/master/data/AmesHousing.txt"
df_ames = pd.read_csv(url, sep="\t")
```

df_ames

```python
df_ames[10:20]
```

```python
print("Total Row count",df_ames['Order'].count())
```

```python
# Linear Interpolation, NaN data replaced with string "Missing"
df_ames = df_ames.interpolate(method="linear")
x = df_ames.isnull().sum()
missing = 0
for (data) in x:
    if data != 0:
        missing+=1
        
print("Number features missing values:", missing)
df_ames.fillna("Missing",inplace=True)
```

```python
#for (data) in 
df_ames[10:20]
```

```python
# Standardization
from sklearn import preprocessing

standardized_df = pd.DataFrame(df_ames, columns=['MS SubClass','Lot Frontage','Lot Area','Overall Qual','Overall Cond',
                                                      'Year Built','Year Remod/Add','Mas Vnr Area','BsmtFin SF 1','BsmtFin SF 2',
                                                      'Bsmt Unf SF','Total Bsmt SF','1st Flr SF','2nd Flr SF', 'Low Qual Fin SF',
                                                      'Gr Liv Area','Bsmt Full Bath','Bsmt Half Bath','Full Bath','Half Bath',
                                                      'Bedroom AbvGr','Kitchen AbvGr','TotRms AbvGrd','Fireplaces','Garage Yr Blt',
                                                      'Garage Cars','Garage Area','Wood Deck SF','Open Porch SF','Enclosed Porch',
                                                      '3Ssn Porch','Screen Porch','Pool Area','Misc Val','Mo Sold','Yr Sold','SalePrice'])

standardized_df_ames = preprocessing.scale(standardized_df, with_mean=True)
#standardized_df_ames[:1]
standardized_df_ames = pd.DataFrame(standardized_df_ames, columns=['MS SubClass','Lot Frontage','Lot Area','Overall Qual','Overall Cond',
                                                      'Year Built','Year Remod/Add','Mas Vnr Area','BsmtFin SF 1','BsmtFin SF 2',
                                                      'Bsmt Unf SF','Total Bsmt SF','1st Flr SF','2nd Flr SF', 'Low Qual Fin SF',
                                                      'Gr Liv Area','Bsmt Full Bath','Bsmt Half Bath','Full Bath','Half Bath',
                                                      'Bedroom AbvGr','Kitchen AbvGr','TotRms AbvGrd','Fireplaces','Garage Yr Blt',
                                                      'Garage Cars','Garage Area','Wood Deck SF','Open Porch SF','Enclosed Porch',
                                                      '3Ssn Porch','Screen Porch','Pool Area','Misc Val','Mo Sold','Yr Sold','SalePrice'])
standardized_df_ames[:10]
```

```python
# Method of Standardization helps us to visualize outliers of the data.
sns.boxplot(x=standardized_df_ames['BsmtFin SF 1'],palette='Greens')
```

```python
# Estimating Mutual information between numeric features and class colum "SalePrice"
import sklearn
from sklearn import feature_selection
#u =pd.DataFrame(index=range(df_ames['PID'].count()),columns=range(1))
y = standardized_df_ames['SalePrice']
X = standardized_df_ames.drop(columns=['SalePrice'])
# print(u)
result = sklearn.feature_selection.mutual_info_regression(X,y)
result_names = X.columns

mutual_info = pd.DataFrame({'Attribute': result_names, 'Mutual info': result})

print("Top 5 features ranked by mutual information: \n\n",mutual_info.sort_values(by=['Mutual info'],ascending=False)[:5])

#print(result, result_names)
```

The top 5 features show that price is most affected by Overall home Quality, Greater Living Area, Garage Area, Year Built and Total basement square footage. Based on this I would expect that newer homes or homes that have been greatly maintained has increased sales prices. Not surprisingly this is expected.
